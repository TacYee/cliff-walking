```python
runs=100 #number of experiments
episodes=500
```

## epsilon=0

```python
reward_perrun_q=[] #sum of q-learning rewards per experiment
reward_perrun_sarsa=[] #sum of sarsa rewards per experiment
```

```python
# q_learning
    for run in range(runs):
        epsilon=0
        reward_episodes= []
        #initialize q table 
        q_table_init() 
        
        for episode in range(episodes):
            reward_steps=0 
            state=(0, 0)
        
            while True:
                state, reward, end_flag=q_learning(state,epsilon)
                reward_steps+=reward
                
                if end_flag:
                    reward_episodes.append(reward_steps)
                    break

            reward_perrun_q.append(reward_episodes) 

        print("***q_learning,experiment %d is completed***"%run)
```

```python
#q-learning path 
OptimalPath(q_table)
```

```python
# sarsa_learning 
for run in range( runs ):
    epsilon=0
    reward_episodes= []     
    #initialize q table 
    q_table_init()    
        
    for episode in range(episodes):
        reward_steps=0
        state=(0, 0)
        
        while True:
            state, reward, end_flag=sarsa_learning(state,epsilon)
            reward_steps+=reward
                
            if end_flag:
                reward_episodes.append(reward_steps)
                break
        reward_perrun_sarsa.append(reward_episodes)

    print("***sarsa_learning,experiment %d is completed***"%run)    
```

```python
#sarsa path
OptimalPath(q_table)
```

```python
import matplotlib.pyplot as plt

points_q=np.array(reward_perrun_q).mean(axis=0)
points_sarsa=np.array(reward_perrun_sarsa).mean(axis=0)
plt.plot(points_q, 'r', label="q-learning",alpha=0.8)
plt.plot(points_sarsa, 'b', label="sarsa-learning",alpha=0.8,linewidth=0.8)
plt.legend(loc = 0)
plt.xlabel('episode')
plt.ylabel('reward sum per episode')
plt.xlim(0, episodes)
plt.ylim(-100, 0)
plt.title("SARSA & Q-LEARNING,epsilon=0")    
plt.show()
```

## Discussion

Observing the path, we can see that q-learning always makes more dangerous moves to walk closer, and sarsa always makes safer moves to walk farther.

As we can see, when the epsilon is zero, the reward curves of q-learning and sarsa are the same. When the exploration rate is zero, the sampling and updating of q-learning and sarsa both use the greedy strategy. Therefore, when the exploration is zero, the two methods are essentially the same.

We started to increase the value of epsilon. As shown in the figure 'epsilon=0.1', the reward of sarsa converges to a higher value. In addition, the convergence rate is higher than q-learning. Although the q-learning can take a closer route, since the epsilon-greedy strategy is used during the action, the agent has a certain chance of falling off the cliff. However, the sarsa took a more conservative route, which lower the probability of falling off the cliff. Therefore, sarsa has a higher reward value and convergence rate. The reason why q-learning is closer to the cliff than sarsa learning is that q-learning uses the greedy strategy in the update phase, which makes the agent tend to move to a state with a higher Q value regardless of risk.

As the value of epsilon gets closer to one, the reward difference between sarsa and q-learning keeps increasing. This is because the increase in the exploration rate causes the probability increase of the agent using q-learning to fall off the cliff.

When epsilon is one, both q-learning and sarsa will not converge. Because the agent continues to explore, it ignores previous experience and is hard to learn.
